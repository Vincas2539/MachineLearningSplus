### Remover os filtros (r, i, z)

```python
del base['mag_psf_r']
del base['mag_psf_i']
del base['mag_psf_z']
```

```python
base # ao remover os filtros devemos ficarm com 9 colunas
```

### Calcular Cores

```python
colunas = ['mag_psf_u', 'mag_psf_J0378', 'mag_psf_J0395', 
           'mag_psf_J0410', 'mag_psf_J0430', 'mag_psf_g', 
           'mag_psf_J0515', 'mag_psf_J0660', 'mag_psf_J0861']

PredictCores = pd.DataFrame()

for i in range(0, 9):
    for j in range(i+1, 9):
        PredictCores[str(colunas[i] + '-' + colunas[j])] = base[colunas[i]].astype(float) - base[colunas[j]].astype(float)
```

```python
PredictCores # Essa base deve ter 36 colunas se utilizarmos 9 filtros
```

```python
PredictCores.to_csv("E:\IC\BasesNovas\Predict\SplusAllFilteredColors.csv")
```
